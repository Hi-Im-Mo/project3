```python
import pandas as pd
import plotly.express as px
import numpy as np
```

```python
df = pd.read_csv("../data/merge_df.csv")
_2023df = df.loc[df["year"]==2023]
del _2023df["Unnamed: 0"]

endeavorDF = _2023df.loc[_2023df["carrier_name"] == "Endeavor Air Inc."].copy()

EDF_carrierDelay = list(endeavorDF["carrier_delay"]/endeavorDF["arr_delay"])
EDF_weatherDelay = list(endeavorDF["weather_delay"]/endeavorDF["arr_delay"])
EDF_nasDelay = list(endeavorDF["nas_delay"]/endeavorDF["arr_delay"])
EDF_securityDelay = list(endeavorDF["security_delay"]/endeavorDF["arr_delay"])
EDF_aircraftDelay = list(endeavorDF["late_aircraft_delay"]/endeavorDF["arr_delay"])

#print(EDF_carrierDelay, EDF_weatherDelay, EDF_nasDelay, EDF_securityDelay, EDF_aircraftDelay)
print(np.sum([EDF_carrierDelay[0],EDF_weatherDelay[0],EDF_nasDelay[0],EDF_securityDelay[0],EDF_aircraftDelay[0]]))
```

```python
airport = "CLT"
carrier = "Endeavor Air Inc."
def generate_pie(airport, carrier):
    selectedDF = _2023df.loc[_2023df["airport"]==airport]
    data = selectedDF.loc[selectedDF["carrier_name"]== carrier]
    columns = ["carrier_delay","weather_delay","nas_delay","security_delay","late_aircraft_delay"]
    values = [(data[i]/data["arr_delay"]).to_numpy()[0] for i in columns]
    labels = ["Delay by Carrier","Delay by Weather","Delay by NAS","Delay by Security","Delay by Late Aircraft"]

    fig = px.pie(names=labels, values=values, labels=labels)
    return fig
```
