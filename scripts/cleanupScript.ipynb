```python
# Import dependencies
import pandas as pd

# Read in raw data
airline_df = pd.read_csv("../data/unclean_airline_delay.csv")

# Inspect raw file
airline_df.info()

# Clean up NaN values
airline_clean = airline_df.dropna()

# Check work
airline_clean.info()

#Reset index after cleaning
airline_clean = airline_clean.reset_index(drop=True)
```

```python
# Initialize list with the top 25 airport IATA codes
codes_list = ["SLC","ATL","FLL","MIA","MCO","MSP","LAX","LAS","DFW","DEN","CLT","PHL","PHX","BOS","IAH","SAN","SEA","SFO","ORD","BWI",'DTW',"EWR","JFK","LGA","DCA"]
# Create a list of the columns from our airline_clean DataFrame
columns_list = list(airline_clean.columns)
# Initialize an empty DataFrame to store our selected data
top25 = pd.DataFrame({}, columns=columns_list)

# Create a for loop to loop threw the airline_clean DataFrame and only grab the rows that have matching values to any of the top 25 IATA codes
for index, row in airline_clean.iterrows():
    if row['airport'] in codes_list:
        top25.loc[len(top25)] = row

# Reset our index
top25.reset_index(drop=True, inplace=True)
# Uncomment to resave CSV
# top25.to_csv("../data/top25_airport_delays.csv")
# Inspect the work
top25.head()
```
